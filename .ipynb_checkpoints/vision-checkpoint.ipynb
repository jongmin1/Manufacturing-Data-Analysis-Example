```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.morphology import skeletonize
import cv2
import json
from sklearn.svm import SVC
```

```python
file_name ="C:/Users/whdal/dataset/vision/raw/right_data.csv"
vectors = np.zeros((423, 81920))
```

```python
r = open(file_name, mode='r')
lines = r.readlines()
```

```python
print(np.shape(lines))
lines[0]
```

```python
for data in range(0, 423):
    numbers = lines[data].split(',')
    flir_vector = []
    
    for i in range(0, 81920):
        flir_vector.append(float(numbers[i]))
    
    vectors[data, :] = np.transpose(flir_vector)
```

```python
from skimage.morphology import skeletonize
import cv2

processed_vectors = np.zeros((423, 80))

for j in range(0,423):
    flir_image = np.reshape(vectors[j,:],[256,320])
    ret, mask = cv2.threshold(flir_image, 41, 255, cv2.THRESH_BINARY)

    mask = mask/255 # normalization

    skeleton = skeletonize(mask)

    mixed = flir_image*skeleton #skeleton 위치의 픽셀 값 이미지화
    mixed_right = mixed[0:255, 160:319]
    min_thresh = 0

    res = np. where(mixed_right>min_thresh)[0]

    maxn = np.max(res) # 제품 ROI 시작좌표
    minn = np.max(res) - 80 # 제품의 ROI 끝 좌표

    num = 0
    vector = np.zeros((80,1))

    for i in range(minn, maxn):
        ii = i-minn+1
        iii = maxn - ii # 제품의 시작 좌표부터 데이터 값 저장을 위한 인덱스 변경

        index = np.where(mixed_right[iii, 0:319] > 0) [0]
    
        vector[num, 0] = mixed_right[iii, index] # 열 데이터 값 저장
        num = num + 1

    vector = np.transpose(vector)
    processed_vectors[j, 0:80] = vector # 전처리 완료 데이터 저장
    
```

```python
import json
with open('C:/Users/whdal/dataset/vision/raw/right_label.json', 'r') as infile:
    newlist = json.load(infile)

thprint(newlist)
```

```python
thrs_l = 0.8
thrs_h = 1.5

label = np.zeros((len(newlist), 1))
num = 0

for val in newlist:
    if thrs_l < val < thrs_h: label[num] = 1
    else: label[num] = 0
    num = num + 1

label
```

```python
import warnings
warnings.filterwarnings(action='ignore')

from sklearn.svm import SVC

datas = np.hstack((processed_vectors, label))
wrong_guess_stack = 0
```

```python
for i in range(0, 100):
    np.random.shuffle(datas)

    train_x = datas[0:300, 0:80]
    test_x = datas[301:423, 0: 80]
    train_y = datas[0:300, 80:81]
    test_y = datas[301:423, 80:81]

    model = SVC(kernel='rbf')
    model.fit(train_x, train_y)

    s_result = model.predict(test_x)

    wrong_guess = np.sum(np.abs(s_result - np.transpose(test_y)))
    wrong_guess_stack = wrong_guess_stack + wrong_guess

(wrong_guess_stack/100/122)*100
```
